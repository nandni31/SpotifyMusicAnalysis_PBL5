```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

songs = [f"Song {i}" for i in range(1, 51)]
artists = [f"Artist {i}" for i in range(1, 51)]
genres = np.random.choice(["Pop", "Hip-Hop", "Rock", "EDM", "Indie"], 50)
popularity = np.random.randint(40, 100, 50)
danceability = np.round(np.random.uniform(0.4, 0.95, 50), 2)
energy = np.round(np.random.uniform(0.3, 1.0, 50), 2)
tempo = np.random.randint(90, 170, 50)

df = pd.DataFrame({
    "song": songs,
    "artist": artists,
    "genre": genres,
    "popularity": popularity,
    "danceability": danceability,
    "energy": energy,
    "tempo": tempo
})

df.head()
```

```python
df.to_csv("spotify_dataset.csv", index=False)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="genre", palette="viridis")
plt.title("Genre Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="energy", y="popularity", hue="genre", palette="coolwarm")
plt.title("Energy vs Popularity")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['tempo'], bins=10, kde=True)
plt.title("Tempo Distribution")
plt.show()
```

```python
numeric_df = df[["popularity", "danceability", "energy", "tempo"]]

plt.figure(figsize=(8,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()

```
